```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("antropometria-dataset.csv")
df.head()
```

# Average Weight by State

```python
pd.pivot_table(df,
               index ='desc_ent',
               aggfunc = ['mean'],
               values = ['peso'],
               margins = True,
              )
```

# Average Weight by Sex by State

```python
pd.pivot_table(df,
               index ='desc_ent',
               aggfunc = ['mean'],
               values = ['peso'],
               margins = True,
               columns ='sexo'
              )
```

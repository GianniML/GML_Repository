```python
path = 'Canada.xlsx'
df = pd.read_excel(path,
             engine = 'openpyxl',
             sheet_name='Canada by Citizenship',
             skiprows = range(20),
             skipfooter= 2)
```

```python
df.head() #.head() es para especificar la cantidad de rows que queres que muestre
```

```python
df.shape #.shape te dice la cantidad de (rows, columnas)
```

```python
df.rename(columns = {'OdName': 'Country',
                     'AreaName' : 'Continent',
                     'RegName' : 'Region'},
          inplace = True)

#df.rename() estamos remplazando el nombre de las columnas

df.head()
```

```python
df.drop(['Type', 'Coverage', 'AREA', 'REG', 'DEV', 'DevName'],
        axis = 1, 
        inplace = True) 

#df.drop es para eliminar o remplazar columnas
#la variable axis es para determinar si son columnas o renglones 1=rows y 2=columnas

df.head()
```

```python
import seaborn as sns
```

```python
df = pd.read_csv('antropometria-dataset-1.csv')
df.head()
```

```python
df.loc[:, ['peso']].plot(kind ='hist',
          bins=15)

plt.ylabel('count')
plt.xlabel('peso')
plt.show()
```

```python
sns.lmplot(x='sistol',
           y='diastol',
           hue='sexo',
           data = df,
           fit_reg= False,
           #Fit_reg saca la regresion
           #Le cambia el color a los datos de acuerdo a los distintos generos
          )
plt.title("Scatter plot of sistol vs diastol", fontsize = 17)
plt.show
```

```python
sns.boxplot(data = df.loc[:,'cintura'])
plt.xticks(rotation = 45, fontsize = 17)
plt.yticks(fontsize = 17)
plt.title('Box plot representation of cintura', fontsize = 15)
plt.show()
```

```python
sns.lmplot(x='sistol',
           y='diastol',
           data = df.loc[df['sexo']==1,['diastol','sistol']],
           fit_reg= False,
           #Fit_reg saca la regresion
           #Le cambia el color a los datos de acuerdo a los distintos generos
          )
plt.title("Scatter plot of sistol vs diastol", fontsize = 17)
plt.show
```

```python
df.loc[df['sexo']==1,['peso']].plot(kind ='hist', bins=25)

plt.ylabel('count')
plt.xlabel('peso')
plt.show()
```

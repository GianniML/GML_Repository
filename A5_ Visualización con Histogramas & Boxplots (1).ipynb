```python
import seaborn as sns
```

```python
df = pd.read_csv('antropometria-dataset-1.csv')
df.head()
```

```python
sns.boxplot(data = df.loc[:,'edad'])
plt.xticks(rotation = 45, fontsize = 17)
plt.yticks(fontsize = 17)
plt.title('Box plot representation of the entire data', fontsize = 15)
plt.show()
```

```python
df2 = df.loc[:, ['peso', 'cintura']]
df2.plot(kind ='line')

plt.xticks(rotation = 45, fontsize = 17)
plt.yticks(fontsize = 17)
plt.title('Line plot representation of "Peso" and "Cinutra"', fontsize = 15)
plt.show()
```

```python
hist_plot = df['peso'].plot(kind = 'hist')
               
plt.xticks(rotation = 45, fontsize = 17)
plt.yticks(fontsize = 17)
plt.title('Histogram representation of "Peso"', fontsize = 15)
plt.show()
```

```python
sns.lmplot(x = 'peso',
           y = 'cintura',
           data = df,
           fit_reg= False,
           #Fit_reg saca la regresion
           hue= 'sexo'
           #Le cambia el color a los datos de acuerdo a los distintos generos
          )
plt.title("Scatter plot of Peso vs Cintura", fontsize = 17)
plt.show
```

```python
df3= pd.pivot_table(df,
               index ='desc_ent',
               columns = 'sexo',
               values = 'peso',
               aggfunc='median',
               margins= True
              )
df3
```

```python
DatosLine = df3.loc[:, ['All']]
DatosLine.transpose()
DatosLine.plot(kind ='line')

plt.xticks(rotation = 45, fontsize = 17)
plt.yticks(fontsize = 17)
plt.title('Line plot representation of "Peso" and "Cinutra"', fontsize = 15)
plt.show()
```

```python
df4= pd.pivot_table(df,
               index ='desc_ent',
               columns = 'sexo',
               values = 'peso',
               aggfunc='sum',
                    margins= True
              )
df4
```

```python
top10 = df4.drop(['All'],
        axis = 0, 
        inplace = True)
```

```python
top10 = df4.sort_values(['All'], ascending = False, axis = 0)
top10 = top10.head(10)
top10
```

```python
top10.plot(kind='bar', figsize=(10, 6))
plt.xlabel('Estado') 
plt.ylabel('Peso') # add y-label to the plot 
plt.show()
```

### Description of Data Set

It is no doubt that squirrels have always been a very underrated animal... Luckily, the folks over at the *NYC Open Data* have created a page for people to register each time they spot a squirrel in a NYC park! 

This data set contains information regarding all the different sightings of squirrels in NYC parks, including a lot of valuable information such as moans, tail twitches and interactions.

*P.S.
Feel free to record a squirrel sighting if you ever visti NYC at: https://www.thesquirrelcensus.com/*

### Description of Variables

long (double): Longitude

lat (double): Latitude

unique_squirrel_id (string): Identification tag for each squirrel sightings. The tag is comprised of "Hectare ID" + "Shift" + "Date" + "Hectare Squirrel Number."

hectare (string): ID tag, which is derived from the hectare grid used to divide and count the park area. One axis that runs predominantly north-to-south is numerical (1-42), and the axis that runs predominantly east-to-west is roman characters (A-I).

# shift (string):* Value is either "AM" or "PM," to communicate whether or not the sighting session occurred in the morning or late afternoon.

date (double): Concatenation of the sighting session day and month.

### 1. What is the predominant age of squirrels spotted and what is the predominant color by age group?

```python
df.drop(['long', 'lat', 'unique_squirrel_id', 'hectare', 'shift', 'date',
       'hectare_squirrel_number','color_notes', 'location', 'above_ground_sighter_measurement',
       'specific_location', 'running', 'chasing', 'climbing', 'eating',
       'foraging', 'other_activities', 'kuks', 'quaas', 'moans', 'tail_flags',
       'tail_twitches', 'approaches', 'indifferent', 'runs_from',
       'other_interactions', 'lat_long', 'zip_codes', 'community_districts',
       'borough_boundaries', 'city_council_districts', 'police_precincts'],
        axis = 1,
        inplace = True
       )

df.head()
```

```python
CountAge=df['age'].value_counts()
CountAge
```

```python
plt.pie(CountAge)

mylabels = ["Adult", "Juvenile", "NaN"]
mycolors = ["orange", "r", "b"]
plt.pie(CountAge, labels = mylabels, colors = mycolors)
plt.legend(title = "Ages:")
plt.title("Pie Chart of Squirrel Ages")
plt.show() 
```

```python
RelationAgeColor=df.pivot_table(index=['age','primary_fur_color'],aggfunc='size',).plot(kind='bar',figsize=(10,5))

plt.ylabel('Number of Squirrels', fontsize = 17)
plt.xlabel('Age/Primary_Fur_Color', fontsize= 17)

plt.yticks (fontsize = 10)
plt.xticks (fontsize = 10, rotation = 45)

plt.title('Number of Squirrels and their Colors by Age', fontsize=17)
```

### Conlusion Question #1

With the data analyzed one can observe that the majority of the squirrels spotted in NYC parks where mainly adults, adding up to a total of 2568 adult squirrels registerd. When looking for tendencies regarding the primary fur color of squirrels by age group, the bar plot was able to show that for both adult and juvenile squirrels, gray colored fur is the most common in NYC. The tendency of primary fur color seems to be the same for both age groups. 

### 2. What are Squirrels usually seen doing when spotted at the morning and at night ?

```python
df2=pd.read_csv("https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2019/2019-10-29/nyc_squirrels.csv")
```

```python
df2.drop(['long', 'lat', 'unique_squirrel_id', 'hectare', 'date',
       'hectare_squirrel_number', 'age', 'primary_fur_color',
       'highlight_fur_color', 'combination_of_primary_and_highlight_color',
       'color_notes', 'location', 'above_ground_sighter_measurement',
       'specific_location', 'kuks', 'quaas', 'moans', 'tail_flags',
       'tail_twitches', 'approaches', 'indifferent', 'runs_from',
       'other_interactions', 'lat_long', 'zip_codes', 'community_districts',
       'borough_boundaries', 'city_council_districts', 'police_precincts','other_activities'],
        axis = 1,
        inplace = True
       )
```

```python
activities=pd.pivot_table(df2,
               index ='shift',
               values = ['running', 'chasing', 'climbing', 'eating','foraging'],
               aggfunc='sum')
activities
```

```python
activities.plot.barh(figsize=(10,5))

plt.ylabel('Shift', fontsize = 17)
plt.xlabel('Number of Squirrels', fontsize= 17)

plt.title('Number of Squirrels and their Colors by Age', fontsize=17)
```

### Conlusion Question #2

When analyzing the bar graph "Number of Squirrels and their Colors by Age" one can clearly observe that the squirrels spotted are usually foraging regardless of whether they are spotted at night or day. Constantly looking for food is a trait we humans seem to share with squirrels. On another hand, besides foraging, the data also shows that squirrels are more likely to be seen climbing trees in the morning and eating at night.

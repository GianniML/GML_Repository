```python
df.columns = list(map(str,df.columns))
df.columns
#Acabamos de cambiar todas las columnas a tipo string
```

```python
df.set_index('Country', inplace = True)
```

```python
df.drop(['Unnamed: 43', 'Unnamed: 44',
       'Unnamed: 45', 'Unnamed: 46', 'Unnamed: 47', 'Unnamed: 48',
       'Unnamed: 49', 'Unnamed: 50'],
        axis = 1,
        inplace = True
       )
df.head()
```

```python
df['Total']=df.iloc[:,2:].sum(axis=1)
df.head()
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
mpl.style.use('ggplot')
```

```python
print(plt.style.available)
# Estos son todos los estilos que tiene matplot para utilizar
```

```python
Japan_Data = df.loc['Japan','1980':'2013']
Japan_Data.plot()

#Estamos diciendo que imprima los valores de japon entre 1980-2013 y que los imprima.
```

```python
IndAndChi_Data = df.loc[['India','China'],'1980':'2013']
IndAndChi_Data.plot()

#Nos damos cuenta que los ejes están invertidos entonces hay que cambiarlos.
```

```python
nic=IndAndChi_Data.transpose()
nic.plot(figsize=(10,5))

plt.ylabel('Number of Imigrants', fontsize = 17)
plt.xlabel('Years', fontsize= 17)

plt.yticks (fontsize = 10)
plt.xticks (fontsize = 10, rotation = 45)

plt.title('Imigration to Canada from India and China during 1980-2013', fontsize=17)
#Con el .transpose() pudimos cambiar los ejes.
```

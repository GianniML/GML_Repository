```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("adult.csv")
df.head()
```

```python
pivot=pd.pivot_table(df,
               index =['education', 'marital-status'],
               aggfunc = ['median'],
               values = ['age','hours-per-week'],
               columns = ['sex']
              )
```

```python
pivot.to_csv('pivot.csv')
```

```python
from sklearn import linear_model
```

```python
df = pd.read_csv("antropometria-dataset.csv")
df
```

```python
df.dropna(inplace = True)
df
```

```python
y = pd.DataFrame(df['cadera'])
x = pd.DataFrame(df['peso'])

lm= linear_model.LinearRegression()
model_lm = lm.fit(x,y)
model_lm
```

```python
modelA = sm.OLS.from_formula('cadera~peso', data = df) # el modelo es OL (least squares)
resultA = modelA.fit() 
resultA.summary()
```

```python
RESI_A = resultA.resid #para ver todos los resiudos de volada
RESI_A
```

```python
print(np.median(RESI_A))
```

```python
y1 = pd.DataFrame(df['entidad'])
x1 = pd.DataFrame(df['sexo'])

lm2= linear_model.LinearRegression()
model_lm2 = lm2.fit(x1,y1)
model_lm2
```

```python
modelB = sm.OLS.from_formula('entidad~sexo', data = df) # el modelo es OL (least squares)
resultB = modelB.fit() 
resultB.summary()
```

```python
y = pd.DataFrame(df['diastol'])
x = pd.DataFrame(df['sistol'])
x1 = pd.DataFrame(df['edad'])
x2 = pd.DataFrame(df['peso'])
x3 = pd.DataFrame(df['cintura'])

lm3= linear_model.LinearRegression()
model_lm3 = lm3.fit(x,x1,x2,x3,y)
model_lm3
```

```python
modelD = sm.OLS.from_formula('diastol~sistol', data = df) # el modelo es OL (least squares)
resultD = modelD.fit() 
resultD.summary()
```

```python
RESI_D = resultD.resid #para ver todos los resiudos de volada
RESI_D
```

```python
print(np.median(RESI_D))
```

```python
import seaborn as sns
```

```python
df = pd.read_csv('Cartwheeldata.csv')
df.head()
```

```python
sns.lmplot(x = 'Wingspan',
           y = 'CWDistance',
           data = df,
           fit_reg= False,
           #Fit_reg saca la regresion
           hue= 'Gender'
           #Le cambia el color a los datos de acuerdo a los distintos generos
          )
plt.title("Scatter plot of CXDistance vs Wingspan", fontsize = 17)
plt.show
```

```python
sns.swarmplot(x = 'Gender',
             y = 'CWDistance',
             data = df
            )
plt.show()
```

# Box plot of Male data

```python
sns.boxplot(data = df.loc[:, ['Age', 'Height', 'Wingspan', 'CWDistance', 'Score']])
plt.xticks(rotation = 45, fontsize = 17)
plt.yticks(fontsize = 17)
plt.title('Box plot representation of the entire data', fontsize = 15)
plt.show()
```

```python
sns.boxplot(data = df.loc[df['Gender']=='M'
                          , ['Age', 'Height', 'Wingspan', 'CWDistance', 'Score']])
plt.xticks(rotation = 45, fontsize = 17)
plt.yticks(fontsize = 17)
plt.title('Box plot representation of the MALE data', fontsize = 15)
plt.show()
```

### 1. Make a comparison of 'Scores' of 'Male' and 'Female'

```python
fig = plt.figure()
ax1 = fig.add_subplot(3,2,1) #num row, num col, position

sns.boxplot(data = df.loc[df['Gender']=='M',
                          ['Score']
                         ]
           )

ax2 = fig.add_subplot(3,2,2) #num row, num col, position

sns.boxplot(data = df.loc[df['Gender']=='F',
                          ['Score']
                         ]
           )

ax3 = fig.add_subplot(3,2,5) #num row, num col, position

sns.boxplot(data = df.loc[df['Gender']=='M',
                          ['Score']
                         ]
           )

ax4 = fig.add_subplot(3,2,6) #num row, num col, position

sns.boxplot(data = df.loc[df['Gender']=='F',
                          ['Score']
                         ]
           )

ax1.title.set_text('Male box plot')
ax2.title.set_text('Female box plot')
ax3.title.set_text('Male box plot')
ax4.title.set_text('Female box plot')
```

```python
import seaborn as sns
```

```python
df = pd.read_csv('antropometria-dataset.csv')
df.head()
```

```python
sns.boxplot(data = df.loc[:, ['Edad']])
plt.xticks(rotation = 45, fontsize = 17)
plt.yticks(fontsize = 17)
plt.title('Box plot representation of the entire data', fontsize = 15)
plt.show()
```

```python
df2 = df.loc[:, ['peso', 'cintura']]
df2.plot(kind ='line')

plt.xticks(rotation = 45, fontsize = 17)
plt.yticks(fontsize = 17)
plt.title('Line plot representation of "Peso" and "Cinutra"', fontsize = 15)
plt.show()
```

```python
hist_plot = df['peso'].plot(kind = 'hist')
               
plt.xticks(rotation = 45, fontsize = 17)
plt.yticks(fontsize = 17)
plt.title('Histogram representation of "Peso"', fontsize = 15)
plt.show()
```

```python
sns.lmplot(x = 'peso',
           y = 'cintura',
           data = df,
           fit_reg= False,
           #Fit_reg saca la regresion
           hue= 'sexo'
           #Le cambia el color a los datos de acuerdo a los distintos generos
          )
plt.title("Scatter plot of Peso vs Cintura", fontsize = 17)
plt.show
```

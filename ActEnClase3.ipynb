```python
df = pd.read_csv('cwurData.csv')
df.head(1)
```

```python
def create_category(x):
    if (x >= 1) & (x <= 100):
        return 'First Tier'
    elif (x>100) & (x<=200):
        return 'Second Tier'
    elif (x<200) & (x<=300):
        return 'Third Tier'
    return 'Other top university'

df['Rank_level'] = df['world_rank'].apply(lambda y:create_category(y))
df.head(2)
```
